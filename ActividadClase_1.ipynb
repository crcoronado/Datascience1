```python
import pandas as pd

# leer el archivo en una dataframe df
df = pd.read_csv('dob_job_application_filings_subset.csv',low_memory=False)
```

```python
#Conocer el nombre de las columnas que componen el dataset
df.columns
```

```python
#Reemplazamos los espacios por _ y las comillas simples por nada en los nombres de las variables para una mejor 
#lecturabilidad y procesamiento de estas.
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace("'", '' )
df.columns
```

```python
#Renombramos algunos nombres de variables para un mejor entendimiento y simplicidad al momento de utilizarlas en el codigo.

df= df.rename(columns={'City_':'City', 'Borough':'Neighbourhood','Community_-_Board':'Community_Board',
                       'Block':'Tax_Block', 'Lot':'Tax_Lot','Total_Est._Fee':'Total_Fee',
                       'ExistingNo._of_Stories':'Existing_Stories', 'Proposed_No._of_Stories':'Proposed_Stories'})

df.head()
```

```python
#En el analisis de las variables encontramo algunas que tenian valores null o que para nosotros no son relevantes en el analisis 
#como por ejemplo OwnersPhone_#

basicColumns = ['Cluster','OwnersHouse_Street_Name','Owners_House_Number','OwnersPhone_#', 'PC_Filed']

for i, var in enumerate(basicColumns):
    print("The unique values in column " + var + " are")
    temp = str(df[var].unique())
    print(temp)
```

```python
#Eliminacion de variables con campos null y no relevantes para el estuddio.
df.drop(basicColumns, axis=1, inplace=True)
df.head()
```

```python
df.iloc[10:15,0:20]
```

```python
#Visualizacion de todas las columnas que quedaron
pd.options.display.max_columns = 100
df.head()
```

```python
# Con el fin de construir un vector o una uniica variables que agrupara las siguientes,se reemplazo el contenido de cada
#de estas (X) por el nombre de la variable, para luego realizar un concatenar
change = ['Plumbing', 'Mechanical', 'Boiler', 'Fuel_Burning', 'Fuel_Storage', 'Standpipe', 'Sprinkler', 'Fire_Alarm', 'Equipment', 'Fire_Suppression', 'Curb_Cut']



for var in change:
    dic = {'X' : var}
    df[var] = df[var].map(dic).fillna(df[var])


df.iloc[10:15,20:35]
```

```python
#Conocer que contiene una variable en especifico
df.Other_Description.unique()
```

```python
# A Esta variable más adelante se le puede realizar una depuración para agrupar las distintas formas que estan explicadas las descripciones
from collections import Counter
other_counts = Counter(df.Other_Description)
other_counts
```

```python
#Concatenar de las variables anteriores para formar una sola
df['Work_type'] = df[change].agg('-'.join, axis=1)
#pd.options.display.max_columns = 100
#df.head()
```

```python
#Revisar las caracteristicas y tipos de variables que contiene el data, con las transformaciones realizadas
df.info()
```

```python
#Algunas variables estan como texto y deberia de ser numericas, por lo tanto es necesario eliminar el signo $ para luego cambiarla de text a int
df.replace({'\$':''}, regex=True, inplace=True)
df
```

```python
#Descripción de las variables numericas 
df.describe() 
```

```python
df.Initial_Cost = df.Initial_Cost.astype(float)
df.info()
```

```python
import numpy as np
#df = df.replace(r'^\s*$', np.NaN, regex=True)
df.Existing_Dwelling_Units = df.Existing_Dwelling_Units.replace(r'^\s*$', np.NaN, regex=True)
```

```python
import numpy as np

change = ['Initial_Cost','Existing_Dwelling_Units','Proposed_Dwelling_Units','Total_Fee']

for col in change:
    df[col] = df[col].replace(r'^\s*$', np.NaN, regex=True)
    df[col] = df[col].astype(float)

df.info()
```
