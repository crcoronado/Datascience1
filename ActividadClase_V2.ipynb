Una lista de las solicitudes de trabajo presentadas para un día en particular y los datos asociados. Los informes semanales y mensuales anteriores se archivan en el DOB y no están disponibles en los datos abiertos de la Ciudad de Nueva York.

**Nota:** estas solicitudes de trabajo se envían a través de las oficinas municipales, a través de eFiling o a través del HUB. Este conjunto de datos no incluye los trabajos enviados a través de DOB NOW.<br>

```python
import pandas as pd

# leer el archivo en una dataframe df
df = pd.read_csv('dob_job_application_filings_subset.csv',low_memory=False)
```

```python
df.columns
```

```python
#Conocemos el nombre de las columnas
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.replace("'", '' )
df.columns
```

```python
df= df.rename(columns={'City_':'City', 'Borough':'Neighbourhood','Community_-_Board':'Community_Board',
                       'Block':'Tax_Block', 'Lot':'Tax_Lot','Total_Est._Fee':'Total_Fee',
                       'ExistingNo._of_Stories':'Existing_Stories', 'Proposed_No._of_Stories':'Proposed_Stories'})

df.head()
```

```python
basicColumns = ['Cluster','OwnersHouse_Street_Name','Owners_House_Number','OwnersPhone_#', 'PC_Filed']

for i, var in enumerate(basicColumns):
    print("The unique values in column " + var + " are")
    temp = str(df[var].unique())
    print(temp)
```

```python
df.drop(basicColumns, axis=1, inplace=True)
df.head()
```

```python
change = ['Plumbing', 'Mechanical', 'Boiler', 'Fuel_Burning', 'Fuel_Storage', 'Standpipe', 'Sprinkler', 'Fire_Alarm', 'Equipment', 'Fire_Suppression', 'Curb_Cut']



for var in change:
    dic = {'X' : var}
    df[var] = df[var].map(dic).fillna(df[var])


df.iloc[10:15,20:35]
```

```python
import numpy as np
```

```python

changefloat = ['Initial_Cost','Existing_Dwelling_Units','Proposed_Dwelling_Units','Total_Fee']



for col in changefloat:
    df[col] = df[col].replace(r'^\s*$', np.NaN, regex=True)
    df[col] = df[col].astype(float)



df.info()
```

```python
boolean_list = ['Landmarked','Adult_Estab', 'Loft_Board', 'City_Owned', 'Little_e', 'eFiling_Filed', 'Professional_Cert', 
                'Horizontal_Enlrgmt', 'Vertical_Enlrgmt', 'Non-Profit'] 
 
for col in boolean_list:
    print(df[col].unique())
    df[col].replace(r'^\s*$', np.NaN, regex=True,inplace=True)
    df[col].fillna(False,inplace=True)
    df[col].replace({'N':False},inplace=True)
    df[col].replace({'Y':True},inplace=True)  
    df[col] = df[col].astype(bool)

df.info()
 
for col in boolean_list:
    print(df[col].unique())
```

```python
df.drop('Other_Description', axis=1, inplace=True)
df['Other'].replace({'X':'Other'},inplace=True)
pd.options.display.max_columns = 100
df.head()
```
